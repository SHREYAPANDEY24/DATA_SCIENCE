# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot. 
Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("first graph")
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

subplot() function is used to multiple plots in a single frame.

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(1,4,1)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(1,4,2)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(1,4,3)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(1,4,4)
plt.plot(x,y)
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

a bar plot is a type of plot which is used to show thae relatonship between two parameters which are represented by x-axis and y-axis respectively.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```
